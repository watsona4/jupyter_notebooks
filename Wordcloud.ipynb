```python
from wordcloud import WordCloud 
import matplotlib.pyplot as plt 
```

```python
def nnl_color_func(*args, font_size, random_state, **kwargs):
    main_colors = [(50, 57, 111), (64, 174, 73)]
    all_colors = [(50, 57, 111), (132, 158, 170), (64, 174, 73), (117, 208, 120), (177, 160, 137), (208, 197, 184)]
    colors = main_colors if font_size > 100 else all_colors
    return random_state.choice(colors)
```

```python
comment_words = """
Integrated Software Technologies
Deliver integrated, multi-disciplinary reactor simulation and analysis
capability to the Program
Drive the open source software development model at NNL
Support multiple technical communities software needs and help them transition
from current to new methods and technologies

Integrated Reactor Engineering
Drive development of predictive simulation frameworks that model prototypical
integrated plant behavior. Develop and deliver an extensible framework that
provides a range of fidelity, common (best estimate plus uncertainty based)
design basis methodologies, and a user-friendly interface.

Computational Methods
Drive the quick adoption and application of new computational technologies
that improve engineering throughput and judgement and accelerate engineering
innovation.

Software Engineering
Design, develop, and deploy an integrated, cross-discipline reactor design
and engineering framework providing a common model-building, workflow
development, job execution, and results processing environment for all aspects
of reactor design and fleet support activities.
"""

UPPER = {'NNL'}
TITLE = {'Program'}

def case(wordlist):
    for v in wordlist.split():
        if v in UPPER:
            yield v.upper()
        if v in TITLE:
            yield v.title()
        yield v.lower()

words = ' '.join(case(comment_words))

wordcloud = WordCloud(width=1102*2, height=458*2,
                      background_color=None, 
                      mode='RGBA',
                      min_font_size=10,
                      color_func=nnl_color_func,
                      random_state=5,
                      scale=4).generate(words)
```

```python
plt.figure(figsize = (11.02, 4.58), facecolor = None) 
plt.imshow(wordcloud, interpolation='bilinear') 
plt.axis("off") 
plt.tight_layout(pad = 0) 

plt.show()
```
